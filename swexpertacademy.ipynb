```python
"https://www.swexpertacademy.com/main/solvingProblem/solvingProblem.do"

"""
2048게임

타일들을 밀어서 합치고 2048만드는게 목적
상하좌우를 정해서 민다.
모든 타일이 그 방향으로 밀림.
        -> 같은 숫자를 만난다 : 합쳐짐
        -> 합쳐진 새로운 숫자는 다시 합쳐지지 않음.

1<=N<=20
각 숫자들은 2의 제곱수들
"""

T = int(input())
 
for test_case in range(1, T + 1):
    next_input = input().split()
    N = int(next_input[0]); direction = next_input[1]
    tiles = [input().split() for x in range(N)]
    tiles = [[int(y) for y in x] for x in tiles]
     
    def push(front,back,x,y):
        if front == 0 :
            return back,0
        elif front == back:
            return front*2+1e-4 * (x+1) * (y+1), 0
        else:
            return front,back
 
    def linepush(line):
        for y in range(len(line)-1):
            for x in range(len(line)-1):
                line[x:x+2] = push(line[x],line[x+1],x,y)
        return line
     
    def transpose(tiles):
        return list(map(list,zip(*tiles)))
     
    if direction == "right":
        result = [linepush(line[::-1])[::-1] for line in tiles]
    elif direction == "up":
        result = [linepush(line) for line in transpose(tiles)]
        result = transpose(result)
    elif direction == "down":
        result = [linepush(line[::-1])[::-1] for line in transpose(tiles)]
        result = transpose(result)
    else:
        result = [linepush(line) for line in tiles]
     
    result = [[round(x) for x in line] for line in result]
     
    """
    left : 그대로하면됨
    right : 반대로 바꿔서 linepush하고 반대로바꾸면됨
    up : transpose해서 linepush하고 다시 transpose
    down : transpose하고 반대로바꿔서 linepush하고 반대로바꾸고 다시 transpose
    """
     
    print(f"#{test_case}")
    for x in range(N):
        print(*result[x],sep=" ")
```

```python
"""
현재 성수의 실력 : 0
강좌 N개
K개를 적절한 순서로 시청하고싶음.
근데 당연히 큰순서대로 하면 되지않음?

ex)

K = 2
2 3
-> 2

K = 1
2 3
-> 1.5

K = 3
9 5 2 3 5
-> 3.75 + 9 /2 = 6.375..

"""
T = int(input())

for test_case in range(1,T+1):
    [N,K] = list(map(int,input().split()))
    numbers = list(map(int,input().split()))
    valid_numbers = sorted(numbers,reverse=True)[:K][::-1]
#     print("valid_numbers\n",
#           valid_numbers)
    before_numbers = [0] + valid_numbers[:-1]
#     print("before_numbers\n",
#           before_numbers)
    
    result = 0
    idx = 0
    while(True):
        result = (valid_numbers[idx] + result)/2
        idx += 1
        if idx == K:
            break
            
    print(f"#{test_case} {result:.6f}")
    
```
