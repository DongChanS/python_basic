```python
T = int(input())
line = input().split()
if len(line) != 1:
    A = [int(x) for x in line][1:]
else:
    A = list(map(lambda x : int(x),input().split()))
int(input())
B = list(map(lambda x : int(x),input().split()))[:1000]

A_dict = {x:0 for x in range(1,T)}
B_dict = {x:0 for x in range(1,T)}

def find_pair(nums, dic):
    for start in range(len(nums)):
        sum_start = 0
        for end in range(start,len(nums)):
            sum_start += nums[end]
            if sum_start <= T-1:
                dic[sum_start] += 1
            else:
                break
                
    return dic

A_dict = find_pair(A,A_dict)
B_dict = find_pair(B,B_dict)

"""
합이 t가 되는 A의 부분합 개수를 f(A,t)
합이 t가 되는 B의 부분합 개수를 f(B,t)

라고 하면,

총 개수는
f(A,i)*f(B,T-i)의 합과 같다.

예를 들면,

A = (1,3,1,2)
B = (1,3,2)

f(A,1) = 2
f(A,2) = 1
f(A,3) = 1
f(A,4) = 2

f(B,4) = 1
f(B,3) = 1
f(B,2) = 1
f(B,1) = 1

-> 합은 2*1+1*1+1*1+2*1 = 7

pair을 전부 무식하게 세고,
합을 전부 기록한다음에 딕셔너리에 개수만 넣어서 전부 곱하면될듯.

그런데 합을 효율적으로 계산하는게 중요할듯.
(무식하게 하면 O(n**4)기 때문에 초과날듯)

"""

print(sum([x*y for x,y in zip(A_dict.values(),list(B_dict.values())[::-1])]))
```

```python
"""
1<= N <= 30000

ex) N = 100
두번째 = 60

100 60 40 20 20 0 20

두번째 = 62

100 62 38 24 14 10 4 6

길이가 최대인 수들을 구하는 프로그램 작성하기.

"""

N = int(input())

max_len = 1
max_list = []

for second in range(1,N+1):
    temp_len = 1
    first = N
    temp_list = [first,second]
    while(True):
        first, second = second, first-second
        temp_len += 1
        if second < 0:
            break
        temp_list.append(second)
    if temp_len >= max_len:
        max_len = temp_len
        max_list = temp_list

print(max_len)
print(*max_list,sep=" ")
```
